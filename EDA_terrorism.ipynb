#### Top 10 Countries with maximum terrorist attacks

```python
#Number of kills grouped by country
c_kills=df[["Country","Killed"]].groupby("Country").sum().sort_values(by="Killed",ascending=False)
c_kills=c_kills[:10]
print(c_kills)

#Number of wounded grouped by country
c_wound=df[["Country","Wounded"]].groupby("Country").sum().sort_values(by="Wounded",ascending=False)
c_wound=c_wound[:10]
print(c_wound)
```

### Terrorist Attacks by the year

```python
yr=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index()
yr.rename(columns={'index':'year','Year':'count'},inplace = True)
yr=yr.set_index('year')
yr.head()
```

```python
yr.plot(kind='bar',color='c',figsize=(15,6))
plt.title('Terrorist attacks by the year',weight='bold')
plt.xlabel('Years',weight='bold')
plt.ylabel('Attack Count',weight='bold')
plt.show()
```

#### From the bar graph above, it is observed that a maximum of 16903 terrorist attacks occured in the year 2014

```python
df_yr = yr.sort_values(by=['count'], ascending=False)
df_yr['Successful'] = df[df['Success']==1].groupby(df['Year']).size()
df_yr['Not Successful'] = df[df['Success']==0].groupby(df['Year']).size()
df_yr.fillna(0, inplace=True)
df_yr.head()
```

```python
df_yr[0:10].plot(kind='bar',color=('g','b','y'),figsize=(15,6))
plt.title(' Yearly Attacks',weight='bold',fontsize=16)
plt.xlabel('Year',weight='bold',fontsize=14)
plt.ylabel('Attack Count',weight='bold')
plt.show()
```

### Casualties caused by Terrorism groups

```python
effect=pd.DataFrame(df[["Group Name","Killed"]].groupby("Group Name").sum().sort_values(by="Killed",ascending=False).drop("Unknown"))
effect['Wounded']=df[["Group Name","Wounded"]].groupby("Group Name").sum().drop("Unknown")
effect.head(10)
```

```python
effect[0:10].plot(kind='bar',color=('c','m'),figsize=(15,6))
plt.title('Casualties caused by Terrorist groups',weight='bold',fontsize=16)
plt.xlabel('Terrorist Group',weight='bold',fontsize=14)
plt.ylabel('Casualty Count',weight='bold')
plt.show()
```

```python
top10 = df.groupby('Group Name')['Year'].count().nlargest(11).index
df_top = df[df['Group Name'].isin(top10)]
df_top = df_top[~(df_top['Group Name']=='Unknown')]
```

```python
df_temp = pd.DataFrame(df_top[df_top['Success']==1].groupby(df_top['Group Name']).size(), columns=['Successful'])
df_temp['Not_Successful'] = df_top[df_top['Success']==0].groupby(df_top['Group Name']).size()
df_temp.reset_index(inplace=True)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(16,8))
sns.barplot(y='Group Name', x='Successful', data=df_temp, color='m', label='Successful')
sns.barplot(y='Group Name', x='Not_Successful', data=df_temp, color='c', label='Failure')
#plt.yscale('log')
#plt.xticks(rotation=90)
plt.ylabel('Terrorist Groups', weight='bold')
plt.xlabel('Number of Attacks', weight='bold')
plt.legend(title='Success', shadow=True, facecolor='lightyellow')
plt.show()
```

```python
df_temp['Success_rate'] = df_temp["Successful"]/(df_temp['Successful']+df_temp['Not_Successful'])
df_temp.reset_index(inplace=True)
```

```python
df_temp
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(16,8))
sns.barplot(y=df_temp['Group Name'], x=df_temp['Success_rate'], palette='husl')
plt.ylabel('Group', weight='bold')
plt.xlabel('Success Rate', weight='bold')
plt.title('Terrorist Groups Success Rate', weight='bold')
plt.show()
```

```python
pd.crosstab(df_top['Group Name'],df_top['Attack Type']).plot.barh(stacked=True, figsize=(16,8))
plt.title('Top 10 Groups based on Attack Type', weight='bold')
plt.xlabel('Number of Attacks', weight='bold')
plt.ylabel('Terrorist Group', weight='bold')
```

```python
pd.crosstab(df_top['Group Name'],df_top['Weapon']).plot.barh(stacked=True, figsize=(16,8))
plt.title('Top 10 Groups based on Weapon', weight='bold')
plt.xlabel('Number of Attacks', weight='bold')
plt.ylabel('Terrorist Group', weight='bold')
```
